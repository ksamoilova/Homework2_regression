```python
import pandas as pd
```

```python
df = pd.read_csv ('homework.csv')
```

crim
per capita crime rate by town.

zn
proportion of residential land zoned for lots over 25,000 sq.ft.

indus
proportion of non-retail business acres per town.

chas
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

nox
nitrogen oxides concentration (parts per 10 million).

rm
average number of rooms per dwelling.

age
proportion of owner-occupied units built prior to 1940.

dis
weighted mean of distances to five Boston employment centres.

rad
index of accessibility to radial highways.

tax
full-value property-tax rate per $10,000.

ptratio
pupil-teacher ratio by town.

black
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

lstat
lower status of the population (percent).

medv
median value of owner-occupied homes in $1000s.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.svm import SVR
```

```python
df.columns
```

```python
X, y = df[['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']], df['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head()
```

```python
y_train.head()
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
X_train.head()
```

### Linear Regression

```python
linear = LinearRegression()
```

```python
linear.fit(X_train, y_train)
```

```python
print('intercept:', linear.intercept_)
print('slope:', linear.coef_)
```

```python
linear.predict(X_train)
```

#### SVR

```python
svr_ = SVR()
svr_.fit(X_train, y_train)
```

```python
svr_.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

### метрики для Linear Regression

```python
mean_squared_error (y_test, linear.predict(X_test))
```

```python
r2_score (y_test, linear.predict(X_test))
```

```python
mean_absolute_error (y_test, linear.predict(X_test))
```

### метрики для SVR 

```python
mean_squared_error (y_test, svr_.predict(X_test))
```

```python
r2_score (y_test, svr_.predict(X_test))
```

```python
mean_absolute_error (y_test, svr_.predict(X_test))
```
